```python
# Instance Method
# A method that operates on an instance of a class and can modify object attributes.
class Example:

  def __init__(self, value):
    self.value = value
  def show_result(self):
    return self.value
example =Example(40)
print(example.show_result())
```

```python
# Example 2: area – Instance method that returns a value
class Circle:
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius ** 2

c1 = Circle(5)
print("Area:", c1.area())
```

```python
# Example 3 :  change_password – Update instance variable
class User :

  def __init__(self , username, password):
    self.username = username
    self.password = password
  def change_password(self , new_password):
    self.password = new_password
ur = User("ninad" , "abc@123")
ur.change_password("xyz@abc")
print(ur.password)
```

```python
# Class Method : -A method that operates on the class rather than on instance objects.
# Uses @classmethod decorator and takes a cls as the first parameter.
```

```python
class Example :
  count = 0

  @classmethod
  def increment_count(cls):
    cls.count +=1
    return cls.count
print(Example.increment_count())

```

```python
# ex 2: school_name – Accessing a class variable

class student :

  school_name = "abc high school"

  @classmethod
  def get_school(cls):
    return cls.school_name
print (student.get_school())
```

```python
#  Example 3: set_raise_amount – Modify class variable

class Employee:
    raise_amount = 1.05

    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    @classmethod
    def set_raise_amount(cls, amount):
        cls.raise_amount = amount

Employee.set_raise_amount(1.10)
print(Employee.raise_amount)
```

```python
class Product:
    total_products = 0

    def __init__(self, name):
        self.name = name
        Product.total_products += 1

    @classmethod
    def info(cls):
        print("Total Products:", cls.total_products)

p1 = Product("Laptop")
p2 = Product("Phone")
Product.info()
```

```python
#Creating a class method
class MyClass:
    count = 0

    def __init__(self):
        MyClass.count += 1

    @classmethod
    def get_count(cls):
        return cls.count

# Creating instances
obj1 = MyClass()
obj2 = MyClass()
obj3 = MyClass()

# Calling the class method
print(MyClass.get_count())  # Output: 3
```

```python
# Static Method
#A method that does not modify the class or instance attributes.
# •Uses @staticmethod decorator and does not require self or cls.
```

```python
class mathoperation :
  @staticmethod
  def add(a,b):
    return a+b
print(mathoperation.add(10,20))
```

```python
# Example 1: greet – Simple static method

class Helper:
    @staticmethod
    def greet():
        print("Hello! Welcome to Python.")

Helper.greet()
```

```python
#  Example 2: is_even – Utility checker
class math:
  @staticmethod
  def is_even(num):
   return num % 2 == 0
print(math.is_even(10))

```
