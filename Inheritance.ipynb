```python

# Inheritance is a fundamental concept in object-oriented programming (OOP) that allows a class (called a child or derived class) to inherit attributes and methods from another class (called a parent or base class).# Types of Inheritance:

# # Single Inheritance: A child class inherits from a single parent class.

# # Multiple Inheritance: A child class inherits from multiple parent classes.

# # Multilevel Inheritance: A child class inherits from another child class, creating a chain of inheritance.

# # Hierarchical Inheritance: Multiple child classes inherit from a single parent class.

# # Hybrid Inheritance: A combination of different inheritance types.
```

```python
# ex .1 . Single Inheritance
class car():
  @staticmethod
  def start():
    print("car started")

  @staticmethod
  def stop():
    print("car stopped")

class toyota(car):
  def __init__(self,name):
    self.name = name

car1 = toyota("corolla")
car1.start()
```

```python
# # Multilevel Inheritance
class car():
  @staticmethod
  def start():
    print("car started")

  @staticmethod
  def stop():
    print("car stopped")

class toyota(car):
  def __init__(self,name):
    self.name = name

class innova(toyota) :
  def __init__(self,type ):
    self.type = type

car1=innova("petrol")
car1.start()
```

```python
# Multiple Inheritance

class A:
  varA = "welcome to claa A"
class B:
  varB = "welcome to claa B"
class C(A,B):
  varC = "welcome to claa C"

c1 = C()
print(c1.varA)
print(c1.varB)
print(c1.varC)

```

```python
# super method
#super() function is used to call the parent class’s methods. In particular, it is commonly used in the child class’s __init__() method to initialize inherited attributes.

# This way, the child class can leverage the functionality of the parent class.

# super() method is used to access method of the paresnt class

class car():

  def __init__(self ,type):
    self.type = type

  @staticmethod
  def start( ):
    print("car started")

  @staticmethod
  def stop():
    print("car stopped")

class toyotaCar(car):
  def __init__(self,name,type):
    super().__init__(type)
    super().start()
    self.name = name

car1 = toyotaCar("corolla","electric")
car1.start()
print(car1.type)
```

```python
# ex . 2
# Parent Class: Person
class Person:
    def __init__(self, name, idnumber):
        self.name = name
        self.idnumber = idnumber

    def display(self):
        print(self.name)
        print(self.idnumber)

# Child Class: Employee
class Employee(Person):
    def __init__(self, name, idnumber, salary, post):
        super().__init__(name, idnumber)  # Using super() to call Person's __init__()
        self.salary = salary
        self.post = post
emp1 =Employee("Ninad" ,"123345", 20000 , "engineer")
emp1.display()
print(emp1.salary)
```

```python
# property() function
# property() function in Python is a built-in function that returns an object of the property class. It allows developers to create properties within a class, providing a way to control access to an attribute by defining getter, setter and deleter methods.

class Student :
  def __init__(self ,phy ,chem ,mat):
    self.phy= phy
    self.chem = chem
    self.mat= mat
  @property
  def percentage(self):
    return str((self.phy + self.chem + self.mat)/3)+"%"
stu1 = Student(90,80,78)
print(stu1.percentage)
stu1.phy = 98
print(stu1.percentage)
```
