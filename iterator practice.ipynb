```python
# 1. Create a Custom Iterator for Counting Implement a class Counter that behaves like an iterator, counting from 1 to n.

class Counter :

  def __init__(self,n):
    self.n = n
    self.current = 1

  def __iter__(self):
    return self

  def __next__(self):
    if self.current < self.n :
      val =self.current
      self.current +=1
      return val
    else :
      raise StopIteration

counter = Counter(6)
for num in counter:
      print(num)
```

```python
# 3. Implement an Infinite Iterator . Create an iterator InfiniteCounter(start) that starts at start and increments indefinitely.
class InfiniteCounter:
    def __init__(self, start):
        self.current = start - 1

    def __iter__(self):
        return self

    def __next__(self):
        self.current += 1
        return self.current

infinite = InfiniteCounter(5)
for _ in range(5):  # Print first 5 values
    print(next(infinite))
```

```python
# 6. Create an Iterator for Squares of Numbers . Write an iterator SquareIterator(n) that yields squares of numbers from 0 to n-1.

class SquareIterator:
    def __init__(self, n):
        self.n = n
        self.current = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.current < self.n:
            square = self.current ** 2
            self.current += 1
            return square
        else:
            raise StopIteration

squares = SquareIterator(5)
print(list(squares))
```

```python
# 4. Implement a String Iterator .Write a class StringIterator that iterates over characters in a string one at a time.

class StringIterator:
    def __init__(self, word):
        self.word = word
        self.index = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.index < len(self.word):
            char = self.word[self.index]
            self.index += 1
            return char
        else:
            raise StopIteration

it = StringIterator("hello")
print(list(it))
```

```python
# 7. Implement a Reverse Iterator for a List Create ReverseIterator(lst), which iterates over a list in reverse order.

class ReverseIterator:
    def __init__(self, lst):
        self.lst = lst
        self.index = len(lst)

    def __iter__(self):
        return self

    def __next__(self):
        if self.index > 0:
            self.index -= 1
            return self.lst[self.index]
        else:
            raise StopIteration

rev = ReverseIterator([1, 2, 3, 4])
print(list(rev))
```

```python
# 8. Implement a Fibonacci Iterator Create a class FibonacciIterator(n), which iterates over the first n Fibonacci numbers.
class FibonacciIterator:
    def __init__(self, n):
        self.n = n
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < self.n:
            value = self.a
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return value
        else:
            raise StopIteration

fib = FibonacciIterator(6)
print(list(fib))
```

```python
# 9. Create an Iterator for the Multiplication Table
class MultiplicationTableIterator:
    def __init__(self, num, n):
        self.num = num
        self.n = n
        self.current = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.current < self.n:
            self.current += 1
            return self.num * self.current
        else:
            raise StopIteration

table = MultiplicationTableIterator(3, 5)
print(list(table))
```

```python

# 10. Implement a Range-like Iterator
class CustomRange:
    def __init__(self, start, stop, step):
        self.current = start
        self.stop = stop
        self.step = step

    def __iter__(self):
        return self

    def __next__(self):
        if self.current < self.stop:
            value = self.current
            self.current += self.step
            return value
        else:
            raise StopIteration

custom_range = CustomRange(1, 10, 2)
print(list(custom_range))
```
