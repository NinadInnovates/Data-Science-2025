```python
# 2. Check Palindrome
# Write a Python function that checks if a given string is a palindrome (i.e., the string reads the same backward as forward). Consider ignoring spaces, punctuation, and case sensitivity.
def is_palindrome(text):

    text = text.lower()
    cleaned = ''
    for char in text:
        if char.isalnum():
            cleaned += char
    # Check if the cleaned string is equal to its reverse
    return cleaned == cleaned[::-1]

# Example usage:
print(is_palindrome("A man, a plan, a canal: Panama"))  # Output: True
print(is_palindrome("Hello"))                           # Output: False
```
