```python
# Introduction to Lambda Functions
# • A lambda function is a small anonymous function in Python.
# • It can take multiple arguments but only contains a single
# expression.

# • Syntax:
# • lambda arguments: expression

square  = lambda x :x**2
print(square(5))
```

```python
# Add 10 to a number:
result = lambda x,y : x + y
print(result(10,30))

# Multiply two arguments:
result = lambda x,y : x * y
print(result(10,30))

#• Summarize three arguments:
result =lambda x,y,z : x+y+z
print(result(2,4,6))

# Double a number:
result = lambda x:x*2
print(result(24))
```

```python
# 3.	Calculate the square: Create a lambda function to return the square of a given number.
square_number = lambda x: x**2
print(square_number(10))
```

```python
# 4.	Check if a number is even: Write a lambda function that returns True if a number is even, and False otherwise.
check_even = lambda x: x%2 == 0
print(check_even(10))
```

```python
# 5.	Find the greater number: Create a lambda function that returns the greater of two numbers.
greater_number = lambda x,y : x if x> y else y
print(greater_number(10,20))
```

```python
# 6.	Add 15 to a number: Write a lambda function that adds 15 to a given number.
add_number = lambda x:x+15
print(add_number(10))
```

```python
# 7.	Multiply two arguments: Create a lambda function that multiplies argument x with argument y and prints the result.
multi_number = lambda x,y: x*y
print(multi_number(10,20))
```

```python
# 8.	Check for a specific starting letter: Write a lambda function to check if a given string starts with the letter "C".
string_start = lambda x:x.startswith("N")
print(string_start("Ninad"))
```

```python
# 9.	Extract the last character: Create a lambda function that takes a string and returns its last character.
result= lambda x : x[-1]
print(result('ninad'))

```

```python
# 10.	Sort by the last character: Write a lambda function that sorts a list of strings by the last character.
strings = ["apple", "banana", "cherry", "date"]
sorted_by_last = sorted(strings, key=lambda s: s[-1])
print(sorted_by_last)
```

```python
# 11.	Convert Celsius to Fahrenheit: Create a lambda function that converts Celsius to Fahrenheit.
c_to_f = lambda c: (c * 9/5) + 32
print(c_to_f(99))
```

```python
# 12.	Calculate area of a rectangle: Write a lambda function to calculate the area of a rectangle given its length and width.
area_rectangle = lambda l,b: l*b
print(area_rectangle(10,20))
```

```python
# 13.	Check if a number is divisible by another number: Create a lambda function that checks if one number is divisible by another.
divisible_by = lambda x,y : x/y == 0
print(divisible_by(10,10))
```

```python
# 14.	String reversal: Write a lambda function to reverse a string.
reverse_string = lambda x:x[::-1]
print(reverse_string("ninad"))
```
