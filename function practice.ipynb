```python
# Variable-Length Arguments (*args)
# . Allows passing multiple positional arguments to afunction.
# • Collected as a tuple inside the function.
```

```python
def add_number(*args):
  return sum(args)
print(add_number(1,2,3,4))
```

```python
#1. Summing Numbers
def sum_all(*args):
    total = 0
    for num in args:
        total += num
    return total

print(sum_all(1, 2, 3))
print(sum_all(1, 2, 3, 4, 5))
```

```python
#2. Finding the Minimum
def my_min(*args):
    result = args[0]
    for num in args:
        if num < result:
            result = num
    return result

print(my_min(4, 5, 6, 7, 2))  # Output: 2
print(my_min(10, 2, 8))       # Output: 2
```

```python
#4. Multiplication
def multiply_all(*args):
    result = 1
    for num in args:
        result *= num
    return result

print(multiply_all(2, 3, 4))
```

```python
#Variable-Length Keyword Arguments(**kwargs)
# • Allows passing multiple keyword arguments as a dictionary (dict).
```

```python
def my_function(**kwargs):
       for key, value in kwargs.items():
           print(f"{key}: {value}")

my_function(name="Alice", age=30, city="New York")
```

```python
#2. Combining with Positional and *args
def my_function( *args, **kwargs):
       print("args:", args)
       print("kwargs:", kwargs)

my_function(1, 2, 3, name="Alice", age=30)
```
