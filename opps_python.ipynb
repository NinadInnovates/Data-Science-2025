```python
class Account :

  def __init__(self,bal,acc):
    self.balance=bal
    self.account=acc

  def debit(self , amount):
    self.balance -= amount
    return self.balance

  def credit(self,amount):
    self.balance += amount
    return self.balance

  def get_balance(self):
    return self.balance

acc1 = Account(1000 , 12345)
acc1.credit(2000)
acc1.debit(500)
print(acc1.get_balance())



```
